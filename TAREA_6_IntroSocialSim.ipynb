```python
# names of players
names=['Jim','Jane','Peter','Zoe']
```

```python
# setting up players
society=[{'name':n,'score':0,'strategy':None} for n in names]
```

```python
# each player a dict:
#society
```

```python
strategies=['Rock','Paper','Scissors']
```

```python
payoff={('Rock','Paper'):(0,1),
        ('Paper','Rock'):(1,0),
        ('Rock','Scissors'):(1,0),
        ('Scissors','Rock'):(0,1),
        ('Paper','Scissors'):(0,1),
        ('Scissors','Paper'):(1,0),
        ('Rock','Rock'):(0,0),
        ('Paper','Paper'):(0,0),
        ('Scissors','Scissors'):(0,0)}
```

```python
from random import choice
```

```python
import itertools  # Para generar combinaciones de jugadores
import time  # Para pausar entre rondas
```

## What code would you add to see the live the results of this last tournament?

```python
import pandas as pd
# several rounds
for aRound in range(100):
  print(f"--- Round {aRound} ---")
    # en each round:
  for player1,player2 in itertools.combinations(society,2):
        # each chooses strategy
      player1['strategy']=choice(strategies)

      player2['strategy']=choice(strategies)

        # result from strategy chosen
      result=payoff[player1['strategy'],player2['strategy']]

        # update scores
      player1['score']+=result[0]
      player2['score']+=result[1]

# Agregamos el código para ver los resultados en vivo después de cada ronda
  socialResults = pd.DataFrame(society)
  print("\nCurrent standings:")
  print(socialResults[['name', 'score']].sort_values(by='score', ascending=False))
  print("\n")
  time.sleep(1)  # Pausa de 1 segundo para actualizaciones en vivo
```

```python
# Nombres de los jugadores: los Bajitos + politólogos reconocidos
names = ['Gian', 'Lucho', 'Joel', 'Chavi', 'Carlos Matus', 'Norbert Lechner',
         'Guillermo O’Donnell', 'Fernando Henrique', 'Julio Cotler', 'Alain Touraine']

# Reiniciamos la sociedad
society = [{'name': n, 'score': 0, 'strategy': None} for n in names]

# Simulamos 100 rondas
for aRound in range(100):
    # Enfrentamos a todos los pares de jugadores
    for player1, player2 in itertools.combinations(society, 2):
        # Cada jugador selecciona una estrategia
        player1['strategy'] = choice(strategies)
        player2['strategy'] = choice(strategies)

        # Calculamos el resultado del enfrentamiento
        result = payoff[player1['strategy'], player2['strategy']]

        # Actualizamos los puntajes de ambos jugadores
        player1['score'] += result[0]
        player2['score'] += result[1]

    # Mostramos los resultados en vivo después de cada ronda
    socialResults = pd.DataFrame(society)
    print(f"\n--- Current standings after round {aRound + 1} ---")
    print(socialResults[['name', 'score']].sort_values(by='score', ascending=False))
    print("\n")
    time.sleep(0.5)  # Pausa de 0.5 segundos para simular tiempo en vivo
```

## Abstraction of agents


```python
# Clase base: Player
class Player:
    def __init__(self, name, score=0):  # Inicializamos con nombre y puntaje
        self.name = name
        self.score = score

    def increase_score(self, value):  # Método para incrementar el puntaje
        self.score += value

# Clase mejorada: PlayerBetter
class PlayerBetter(Player):
    def __init__(self, name, score=0):  # Hereda de Player y agrega atributos
        super().__init__(name, score)  # Inicializa usando la clase base
        self.strategy = None

    def get_strategy(self):  # Método para elegir estrategia
        from random import choice
        strategies = ['Rock', 'Paper', 'Scissors']
        self.strategy = choice(strategies)
        return self.strategy

# Clase para manejar el juego
class Game:
    def __init__(self, player_names):  # Inicializamos el juego con nombres de jugadores
        self.players = [PlayerBetter(name) for name in player_names]  # Creamos objetos PlayerBetter

    def play_round(self):  # Jugar una ronda
        import itertools
        payoff = {  # Reglas del juego
            ('Rock', 'Paper'): (0, 1),
            ('Paper', 'Rock'): (1, 0),
            ('Rock', 'Scissors'): (1, 0),
            ('Scissors', 'Rock'): (0, 1),
            ('Paper', 'Scissors'): (0, 1),
            ('Scissors', 'Paper'): (1, 0),
            ('Rock', 'Rock'): (0, 0),
            ('Paper', 'Paper'): (0, 0),
            ('Scissors', 'Scissors'): (0, 0)
        }
        # Enfrentamos a todos los pares de jugadores
        for player1, player2 in itertools.combinations(self.players, 2):
            # Ambos eligen estrategias
            strategy1 = player1.get_strategy()
            strategy2 = player2.get_strategy()

            # Determinamos el resultado
            result = payoff[(strategy1, strategy2)]

            # Actualizamos puntajes
            player1.increase_score(result[0])
            player2.increase_score(result[1])

    def play_game(self, rounds):  # Jugar múltiples rondas
        for _ in range(rounds):
            self.play_round()

    def declare_winner(self):  # Declarar el ganador
        max_score = max(player.score for player in self.players)
        winners = [player.name for player in self.players if player.score == max_score]
        return winners, max_score

    def show_results(self):  # Mostrar resultados finales
        for player in self.players:
            print(f"{player.name}: {player.score}")

# Nombres de los jugadores
player_names = ['Gian', 'Lucho', 'Joel', 'Chavi', 'Carlos Matus', 'Norbert Lechner',
                'Guillermo O’Donnell', 'Fernando Henrique', 'Julio Cotler', 'Alain Touraine']

# Inicializamos el juego
game = Game(player_names)

# Jugamos 100 rondas
game.play_game(100)

# Mostramos los resultados finales
print("\nFinal Results:")
game.show_results()

# Declaramos al ganador
winners, max_score = game.declare_winner()
print(f"\nWinner(s): {', '.join(winners)} with a score of {max_score}!")
```
